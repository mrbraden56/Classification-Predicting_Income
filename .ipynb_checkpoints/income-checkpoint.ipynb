```python
import seaborn as sns
import matplotlib.pyplot as plt
import math
```

```python
def plot_distribution(dataset, cols=5, width=20, height=15, hspace=0.2, wspace=0.5):
    fig = plt.figure(figsize=(width,height))
    fig.subplots_adjust(left=None, bottom=None, right=None, top=None, wspace=wspace, hspace=hspace)
    rows = math.ceil(float(dataset.shape[1]) / cols)
    for i, column in enumerate(dataset.columns):
        ax = fig.add_subplot(rows, cols, i + 1)
        ax.set_title(column)
        if dataset.dtypes[column] == np.object:
            g = sns.countplot(y=column, data=dataset)
            substrings = [s.get_text()[:18] for s in g.get_yticklabels()]
            g.set(yticklabels=substrings)
            plt.xticks(rotation=25)
        else:
            g = sns.distplot(dataset[column])
            plt.xticks(rotation=25)
    
plot_distribution(df, cols=3, width=20, height=20, hspace=0.45, wspace=0.5)
```

This function gives us a good grasp of the graphs for each of the features, showing us its distribution.

Now its time to get to the fun stuff! As of now we have gone very in depth at looking at our data and Visualizing it. Now, its time to actually start manipulating the data. Here are the steps of action we will take...
* We will turn the categorical values into numerical values using pd.get_dummies and OrdinalEncoder.
* Combine the categorical and numeric dataframes after encoding has taken place.
* Remove outliers from dataset.
* Fill the missing values as a result of removing outliers.

We will be using these packages...
* <b>SimpleImputer</b>
* <b>get_dummies</b>
* <b>OrdinalEncoder</b>

to carry out the planned tasks for this section of the project.

Now lets talk about the reasons behind each of the goals we have for this chapter.
1. The machine learning models we will be using do not take categorical data. Computers can only understand numerical values, thus we must convert each categorical value into a numeric value.
2. Combining the the numerical and categorical dataset will make things easier because we will only need to worry about 1 dataset.
3. Removing the outliers will give us a better distribution of values making the machine learning model we will train more accurate.
4. Some columns will have more outliers removed than others. The pandas dataframe does not let columns have different length, so to make up for that, pandas will fill the smaller column with NaN values to make it the same length as the biggest column. We will then need to find a way to deal with the NaN value.

To start, we need to do two things...
* Drop the income column off of categorical_data since income has already been encoded.
* Set numeric_data and categorical_data equal to the entire dataset instead of only x_train, y_train. That way when we go to split the dataset using train_test_split the entire dataset will have been cleaned.

```python
numeric_data=dataframe.select_dtypes(include=[np.number])
categorical_data=dataframe.select_dtypes(exclude=[np.number])
y=dataframe[' income']
```

Next we must look at how we change categorical values into numeric. There are two different types of categorical variables...
* <b>Ordinal</b>
* <b>Nominal</b>

Ordinal is defined as...
* A categorical value in which the order matters. A simply example of this are grades. Grades are as follows, 'A', 'B', 'C', 'D', 'F'. The letter grade 'A' is higher than the letter grade 'C' which makes it ordered.

The way we encode these is through OneHotEncoding or pandas get_dummies which makes a column for each unique value.
 
Nominal is defined as...
* A categorical value in which the order does not matter. An example of this could be different types of animals. If you have a column of Dogs, Cats, and Cows it does not matter what order you put it in.
 
The way we encode these is through LabelEncoding which assings as numerical value to each unique value.

We will start with the nominal features. These are the features where the variable does not have an order.

```python
one_hot=pd.get_dummies(categorical_data[[' marital-status',' relationship',' race',' sex',' native-country']])
```

Next we will do the ordinal features. These are the features where order does matter. We will use the package...
* <b>OrdinalEncoder</b>

which assigns a numeric value for each unique categorical value.

```python
from sklearn.preprocessing import OrdinalEncoder
ordinal=categorical_data[[' workclass',' education',' occupation']]
ordinal=OrdinalEncoder().fit_transform(ordinal)
ordinal=pd.DataFrame(ordinal, columns=[' workclass',' education',' occupation'])
ordinal
```

But before we do that we must make the indexes similar for the dataframes...

* numerical_data and categorical_data_encoded

We must make the one_hot indexes the same as the ordinal indexes for the pandas dataframe concat method to work properly. To do this we made the make_index_similar function.

```python
def make_index_similar(df):
    length=len(df.index)
    index=list(range(0,length))
    df['index']=index
    df.set_index('index', inplace=True)
```

Now the indexes for one_hot and ordinal are the same.

Next, we will be dealing with the outliers in the encoded_df dataframe.

```python
plt.figure(figsize=(3,6))
sns.boxplot(y=encoded_df['age'])
```

Box and whisker plots are a great tool to use when trying to identify outliers. Examing the graph above...
* The bottom horizontal line represents the minimum of the age category.
* In the rectangle, everything below the center line to the bottom of the rectangle is the first quartile. The first quartile is defined as everything to the left of the median. So, if you had numbers 1,2,3,4,5. The median is 3 and the first quartile is 1 and 2.
* The middle line in the rectangle represents the median number of the age category. This is also called the second quartile.
* In the rectangle, everything above the center line to the top of the rectangle is the third quartile. Looking at our example with 1,2,3,4,5. 3 would be the median, second quartile, and  and 5 would be the third quartile.
* The top horizontal line represents the max of the age category.

Now here is the interesting part on how the box and whisker plot relates to outliers. The dots above the maximum horizontal line represents the numbers whose value is more than 3/2 times that of the upper quartile. These are the outliers of this data.

There could also be outliers below the minimum horizontal line. These would be numbers whose value is less than 3/2 times that of the upper quartile.

Now as we saw earlier in our data exploration, it takes a long time! So, instead of graphing every column to see its box and whisker plot, we will create a function that goes through each of the columns, calculates if the number is an outlier, and then remove the outliers from the data.

```python
def find_boundaries(data, variable, distance):
    IQR=data[variable].quantile(0.75)-data[variable].quantile(0.25)
    lower_boundary=data[variable].quantile(0.25)-(IQR*distance)
    upper_boundary=data[variable].quantile(0.75)+(IQR*distance)
    return upper_boundary, lower_boundary
def outliers(df,distance):
    for column in df.columns:
        upper_bound, lower_bound= find_boundaries(df, column, distance)
        outliers=np.where(df[column]>upper_bound, True, np.where(df[column]<lower_bound, True, False))
        df[column]=df.loc[~outliers, column]
        print("Amount of values removed from %s: %d" % (column, df[column].isnull().sum()))
```

```python
outliers(encoded_df, 3.0)
```

Now, looking at the outliers we replaced we NaN we can see the percentage of NaN's per column.

```python
encoded_df.isnull().mean()*100
```

As you can see we now have a lot of missing values in our dataframe. The machine learning algorithms we will be using usually do not work if we have missing values. To solve this problem we will be imputing these missing values by filling them in with the mean of the data set.

We will be using a package called...
* <b>SimpleImputer</b>

```python
from sklearn.impute import SimpleImputer
imputer=SimpleImputer(missing_values=np.nan, strategy='mean')
imputer=imputer.fit_transform(encoded_df)
```

Now we have imputed all the NaN values of the dataframe using the mean!

However, this method returns a numpy array. We want to be using a pandas dataframe so we will convert this numpy array into a pandas dataframe.

```python
columns=list(encoded_df.columns)
```

```python
encoded_df=pd.DataFrame(imputer, columns=columns)
```

Now we can see that the dataframe encoded_df as no missing values.

```python
encoded_df.isnull().sum().sum()
```

# 4. Feature Selection and Reduction

The reason for feature selection and reduction can be summed up by this sentence...
* To reduce the number of features in a dataset to those features which have the most impact on the target feature.

In this sentence two main actions are taking place.
1. Reducing the overall amount of features.
2. Selecting features which have the most impact on the target feature.

You may ask why bother with these two things?

<b>For action number 1.</b>

When machine learning algorthims are fed a dataset in which the number of features are very high they are prone to overfitting. Overfitting is when a machine learning algorthim does predict well. Before we start lets create a function that makes testing these models easier and import everything we need.

```python
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import GradientBoostingClassifier
from xgboost import XGBClassifier

from sklearn.model_selection import cross_val_score
from sklearn.model_selection import cross_val_predict
from sklearn.metrics import roc_curve
from sklearn.metrics import roc_auc_score
from sklearn import metrics


def model(algo, X_train, y_train, cv):
    model = algo.fit(X_train, y_train)
    acc = round(model.score(X_train, y_train) * 100, 2)
    scores=round(cross_val_score(model, X_train, y_train, scoring='accuracy', cv=cv).mean()*100)
    print('Accuracy of model is: %.1f' % acc)
    print('Accuracy of model with cv-10 is: %.1f' % scores)
    
def roc_auc(x_train, x_test, y_train, y_test, model):
    algo=model
    algo.fit(x_train, y_train)
    metrics.plot_roc_curve(algo, x_test, y_test)
```

Now that we have trained our models we can see each models score and roc graphs. But first off, what do these mean?

<b>Score</b>

The score represents the accuracy of the model with a higher score being a more accurate model.

<b>Score with cv-10</b>

The score method uses data that it has seen to calculate the score. This can be misleading because we do not know how it will work with data that it has not seen. To fix this problem we use k-fold cross validation which shuffles the data. This then gives us the accuracy of the model for unseen data.

<b>ROC and AUC</b>

The ROC curve is what you see on the graph and it plots True Positive Rate vs False Positive Rate. True positive rate is defined as the machine learning model classifing something as true when in reality it is true. False positive rate is defined as the machine learning model classifing something as true when in reality it is false. The more the ROC curve is in the top left corner the better.

AUC is defined as area under the curve which calculates the area under the ROC curve. 1 is the best possible score we can get while 0.5 is the score you can get if you try to classify things at random.

Now lets look at all the models on one graph and see which one comes out on top.

```python
log=LogisticRegression()
model_log=log.fit(x_train, y_train)
probs=model_log.predict_proba(x_test)
log_prob=probs[:,1]

K=KNeighborsClassifier()
model_k=K.fit(x_train, y_train)
probs=model_k.predict_proba(x_test)
k_prob=probs[:,1]

svc=SVC(probability=True)
model_svc=svc.fit(x_train, y_train)
probs=model_svc.predict_proba(x_test)
svc_prob=probs[:,1]

rf=RandomForestClassifier()
model_rf=rf.fit(x_train, y_train)
probs=model_rf.predict_proba(x_test)
rf_prob=probs[:,1]

gbc=GradientBoostingClassifier()
model_gbc=gbc.fit(x_train, y_train)
probs=model_gbc.predict_proba(x_test)
gbc_prob=probs[:,1]

xgb=XGBClassifier()
model_xgb=xgb.fit(x_train, y_train)
probs=model_xgb.predict_proba(x_test)
xgb_prob=probs[:,1]

fig=plt.figure(figsize=(10,10))

models=[
    'Logistic Regression',
    'KNN',
    'SVC',
    'Random Forest',
    'Gradient Boosting Classifier',
    'XGBClassifier'
]

probs=[
    log_prob,
    k_prob,
    svc_prob,
    rf_prob,
    gbc_prob,
    xgb_prob    
]

colors = [
    'blue',
    'green',
    'red',
    'cyan',
    'magenta',
    'yellow',
]

plt.title('ROC curves')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.xlim([-0.01, 1.01])
plt.ylim([-0.01, 1.01])

def plot_roc_curves(y_test, prob, model):
    fpr, tpr, threshold=metrics.roc_curve(y_test, prob)
    roc_auc=metrics.auc(fpr, tpr)
    plt.plot(fpr, tpr, 'b', label=model+' AUC=%.2f' % roc_auc, color=colors[i])
    plt.legend(loc='lower right')
    
for i, model in list(enumerate(models)):
    plot_roc_curves(y_test, probs[i], models[i])
    
plt.show()        
```

Looking at the graphs we can see the top two candidates for the most accurate machine learning model is Gradient Boosting Classifier and XGBClassifier.

However, just because these two machine learning models got the best score does not mean that they will come out on top. A lot of these models have parameters that can make there performance better and we have not messed with any of the parameters so far. So, the next step in finding the best machine learning model is to play with each of the parameters for each machine learning model.

Now there are hundreds of combinations and we do not have the time and day to test each one manually. So, what we will be doing is using something called...
* <b>GridSearchCV</b>

which goes through all of the combinations for a specific model and returns the parameters that have the best score. To make this easier we have created a class that can allow us to quickly perform gridsearch on each model.

```python
from sklearn.model_selection import RepeatedStratifiedKFold
from sklearn.model_selection import GridSearchCV

class HyperParameter:
    def __init__(self, x_train, y_train):
        self.x_train=x_train
        self.y_train=y_train
        
        self.cv=RepeatedStratifiedKFold(n_splits=10, n_repeats=3, random_state=1)
        
        self.logistic=LogisticRegression()
        self.kn=KNeighborsClassifier()
        self.svc=SVC()
        self.ridge=RidgeClassifier()
        self.forest=RandomForestClassifier()
        self.gbc=GradientBoostingClassifier()
        self.xgb=XGBClassifier()
    
    def logistic_regression(self, param):
        grid_search=GridSearchCV(estimator=self.logistic, param_grid=param, n_jobs=-1, cv=self.cv, scoring='accuracy', error_score=0)
        grid_result=grid_search.fit(self.x_train, self.y_train)
        print("Best: %f using %s" % (grid_result.best_score_, grid_result.best_params_))
        
    def knn(self, param):
        grid_search=GridSearchCV(estimator=self.kn, param_grid=param, n_jobs=-1, cv=self.cv, scoring='accuracy', error_score=0)
        grid_result=grid_search.fit(self.x_train, self.y_train)
        print("Best: %f using %s" % (grid_result.best_score_, grid_result.best_params_))
        
    def svm(self, param):
        grid_search=GridSearchCV(estimator=self.svc, param_grid=param, n_jobs=-1, cv=self.cv, scoring='accuracy', error_score=0)
        grid_result=grid_search.fit(self.x_train, self.y_train)
        print("Best: %f using %s" % (grid_result.best_score_, grid_result.best_params_))
        
    def ridge_classifier(self, param):
        grid_search=GridSearchCV(estimator=self.ridge, param_grid=param, n_jobs=-1, cv=self.cv, scoring='accuracy', error_score=0)
        grid_result=grid_search.fit(self.x_train, self.y_train)
        print("Best: %f using %s" % (grid_result.best_score_, grid_result.best_params_))
        
    def random_forest(self, param):
        grid_search=GridSearchCV(estimator=self.forest, param_grid=param, n_jobs=-1, cv=self.cv, scoring='accuracy', error_score=0)
        grid_result=grid_search.fit(self.x_train, self.y_train)
        print("Best: %f using %s" % (grid_result.best_score_, grid_result.best_params_))
    
    def gradientboosting_classifier(self, param):
        grid_search=GridSearchCV(estimator=self.gbc, param_grid=param, n_jobs=-1, cv=self.cv, scoring='accuracy', error_score=0)
        grid_result=grid_search.fit(self.x_train, self.y_train)
        print("Best: %f using %s" % (grid_result.best_score_, grid_result.best_params_))
        
    def xgboost(self, param):
        grid_search=GridSearchCV(estimator=self.xgb, param_grid=param, n_jobs=-1, cv=self.cv, scoring='accuracy', error_score=0)
        grid_result=grid_search.fit(self.x_train, self.y_train)
        print("Best: %f using %s" % (grid_result.best_score_, grid_result.best_params_))
        
    def voting_classifier(self, model_list, param, iterations):
        voting_clf=VotingClassifier(
        estimators=model_list,
        voting='hard'
        )
        random_search=RandomizedSearchCV(voting_clf, param, n_iter=iterations, scoring='accuracy', n_jobs=-1, cv=self.cv)
        random_result=random_search.fit(x_train, y_train)
        print("Best: %f using %s" % (random_result.best_score_, random_result.best_params_))
```

<b>Logistic Regression</b>

```python
solvers = ['newton-cg', 'lbfgs', 'liblinear']
penalty = ['l2']
c_values = [100, 10, 1.0, 0.1, 0.01]
grid = dict(solver=solvers,penalty=penalty,C=c_values)

hyperparameter=HyperParameter(x_train, y_train)
hyperparameter.logistic_regression(grid)
```

There we have it! We have done a grid search on just about every machine learning model except XGBoost! The reason for this is since grid search uses a brute force algorithm, meaning it tries all the possible combinations, it take a very long time to compute! Just these 5 models alone took me about 3 hours to compute! That is not realstic in real world scenarios, especially when the datasets are bigger and we use more parameters per model.

To solve this there is something called RandomizedSearchCV. RandomizedSearchCV is the same as GridSearchCV in that you pass the parameters you want into the function. However, RandomizedSearchCV has another parameter called iterations. However many iterations you pass in will be the number of times that RandomizedSearchCV randomly select parameters and tries them out. Then after n iteations it will give you the best result and parameters.

On top of this we will be using an ensemble learning which is when you combine machine learning models to get a better result.

So we will be using...
* <b>RandomizedSearchCV</b>
* <b>VotingClassifier(Ensemble Learning)</b>

```python
from sklearn.model_selection import RandomizedSearchCV
from sklearn.ensemble import VotingClassifier
estimators=[('lr', LogisticRegression()), ('kn',KNeighborsClassifier()), ('gbc', GradientBoostingClassifier())]
#lr
solver=['newton-cg', 'sag', 'lbfgs']
penalty=['l2']
C=[100, 10, 1.0, 0.1, 0.01]
#kn
n_neighbors=range(1,21,2)
weights=['uniform', 'distance']
metric = ['euclidean', 'manhattan', 'minkowski']
#gbc
n_estimators = [10, 100]
learning_rate = [0.001, 0.01, 0.1]
max_depth = [3, 7, 9]

parameters=dict(lr__solver=solver, lr__penalty=penalty, lr__C=[100, 10, 1.0, 0.1, 0.01],
                kn__n_neighbors=n_neighbors, kn__weights=weights, kn__metric=metric,
                gbc__n_estimators=n_estimators, gbc__learning_rate=learning_rate, gbc__max_depth=max_depth)
randomsearch=HyperParameter(x_train, y_train)
randomsearch.voting_classifier(estimators, parameters, 10)
```
